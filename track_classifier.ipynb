```python
import pandas as pd
import numpy as np
import pickle
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

happy_sad_model = pickle.load(open("happy_sad_model.sav", 'rb'))
energetic_calm_model = pickle.load(open("calm_energetic_model.sav", 'rb'))
```

```python
data1 = pd.read_csv("energetic_calm_data.csv",index_col=0)
data2 = pd.read_csv("happy_sad_data.csv",index_col=0)
data = pd.concat([data1, data2])
data = data.drop('instrumentalness', axis = 1)
data.reset_index(inplace=True)
col_features = data.columns[6:-3]
data.head()
```

```python
def classify_songs(track_links):
    #Obtain the features of the song
    ids = [link.split("/")[-1].split("?")[0] for link in track_links]
    #ids = [track_links]
    
    song_data = data.loc[data["id"].isin(ids)]
    X_data = data.loc[data["id"].isin(ids)][col_features].astype(float)
    
    #Predict the features of the song
    for (x, song) in zip(X_data.values, song_data.values):
        calm_energetic_result = str(energetic_calm_model.predict([x]))[2:-2]
        happy_sad_result = str(happy_sad_model.predict([x]))[2:-2]
        song_name = song[0]
        artist = song[2]

        print("{0} by {1} is {2} and {3} song".format(song_name,artist,calm_energetic_result,happy_sad_result))
    
```

```python
data = pd.read_csv('energetic_calm_data.csv')
ids = data['id']
for i in ids:
    classify_songs([i + '?'])
```
